```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
df = pd.read_csv("titanic-passengers.csv", sep=";")
print(df.head())
print("="*100)
print(df.info())
print("="*100)
df.isnull().sum()
```

```python
#CLEANING
df.drop(["Cabin", "PassengerId", "Ticket", "Name", "SibSp", "Parch"], axis=1, inplace=True)
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
print(df.isnull().sum())
```

```python
#TRANSFORMATION
cleanup_dict = {
    "Survived": {"No": 0, "Yes": 1},
    "Sex": {"male": 0, "female": 1},
    "Embarked": {"C": 0, "Q": 1, "S": 2}
}
df.replace(cleanup_dict, inplace=True)

#CORRELATION
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots(figsize = (12 , 10))
    cmap = sns.diverging_palette(220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df)
```

```python
X = df[['Pclass', 'Sex','Age', 'Fare', 'Embarked']]
y = df['Survived']

#SPLITTING
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

# Applying logistic regression

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred  = logreg.predict(X_test)
print("Accuracy = {0:.2f}".format(logreg.score(X_test, y_test)))
```

```python
sns.regplot(x='Age',y='Survived',data=df)
```

# Confusion matrix

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
print(metrics.classification_report(y_test, y_pred))
```

# ROC/AUC

- ROC = Receiver Operating Characteristic

ROC curves typically feature true positive rate on the Y axis, and false positive rate on the X axis. This means that the top left corner of the plot is the “ideal” point - a false positive rate of zero, and a true positive rate of one.


- AUC = Area Under the ROC Curve

AUC measures the entire two-dimensional area underneath the entire ROC curve from (0,0) to (1,1)

```python
#Predict probabilities
p = logreg.predict_proba(X_test)
#Keep probabilities for the positive outcome only
p = p[:, 1]
#Calculate scores                      
roc_auc = metrics.roc_auc_score(y_test, p)
print('ROC/AUC = {0:.2f}'.format(roc_auc))
```

```python
#Calculate roc curves
fpr, tpr, thresholds = metrics.roc_curve(y_test, p)
#Plot the roc curve for the model
plt.plot(fpr, tpr)
```
